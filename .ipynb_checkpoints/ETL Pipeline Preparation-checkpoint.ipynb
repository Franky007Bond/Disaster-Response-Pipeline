### 1. Import libraries and load datasets.

```python
# import libraries
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def df_nan_analysis(df):
    """
        This function analyses both the absolute and relative number of nan elements in the different columns of a dataset
        It generates an overview table by attribute
        Args:
            df: Dataframe to be analysed
            
        Returns:
            df_report: Dataframe containing analysis results:
                - attribute: Attribute name
                - nan_abs: Absolute count of nan-values for this attribute
                - nan_rel: Relative frequency of nan-values for this attribute
      
    """
    df_analysis = df.isnull().sum()
    df_report = pd.DataFrame([])
    for key, value in df_analysis.iteritems():
        df_report = df_report.append({'attribute':key, 'nan_abs':value, 'nan_rel': str(round(100*value/df.shape[0],2))+'%'}, ignore_index=True)
    
    df_report['nan_abs'] = df_report['nan_abs'].astype(int)
    
    return df_report
```

### 3. Split `categories` into separate category columns.
- Split the values in the `categories` column on the `;` character so that each value becomes a separate column. You'll find [this method](https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.Series.str.split.html) very helpful! Make sure to set `expand=True`.
- Use the first row of categories dataframe to create column names for the categories data.
- Rename columns of `categories` with new column names.

```python
# drop rows with nan-values for message or categries-variable
# as these are the arguments that are required for the machine learning model
# this is done to make the algorithm more robust even if the data provided 
# does not have any such issue
df.dropna(subset=['message','categories'])

# create a dataframe of the 36 individual category columns
categories = df['categories'].str.split(pat=';', expand=True)
categories.head()
```

```python
# select the first row of the categories dataframe
row = categories.iloc[0,:]

# use this row to extract a list of new column names for categories.
# one way is to apply a lambda function that takes everything 
# up to the second to last character of each string with slicing
category_colnames = row.apply(lambda x: x[0:-2])
print(category_colnames)
```

```python
# rename the columns of `categories`
categories.columns = category_colnames
categories.head()
```

```python
df_nan_analysis(categories)
```

### 4. Convert category values to just numbers 0 or 1.
- Iterate through the category columns in df to keep only the last character of each string (the 1 or 0). For example, `related-0` becomes `0`, `related-1` becomes `1`. Convert the string to a numeric value.
- You can perform [normal string actions on Pandas Series](https://pandas.pydata.org/pandas-docs/stable/text.html#indexing-with-str), like indexing, by including `.str` after the Series. You may need to first convert the Series to be of type string, which you can do with `astype(str)`.

```python
for column in categories:
    # set each value to be the last character of the string
    categories[column] = categories[column].apply(lambda x: x[-1])
    # alt: categories[column] = categories[column].str[-1]
    
    # convert column from string to numeric
    categories[column] = pd.to_numeric(categories[column])
    # alt: categories[column] = categories[column].astype(int)
    
categories.head()
```

```python
# check the completeness 
df_nan_analysis(categories)
```

```python
categories.max()   
```

**Conclusion**
- At least the column 'related' contains some values of 2 which violates the boolean meaning of the variables
- We need to correct all values of 2 to 1; To make the productio code most robust we shouldn't limit this to one specific column but consider all columns
- the column 'child_alone' doesn't contain any information as it contains zeros only

```python
# correct all values of 2 to 1
for col in categories.columns:
    categories[col] = categories[col].apply(lambda x: 1 if (x==2) else x)
```

### 5. Replace `categories` column in `df` with new category columns.
- Drop the categories column from the df dataframe since it is no longer needed.

```python
# concatenate the original dataframe with the new `categories` dataframe
df = pd.concat([df, categories], axis=1)
#df[categories.columns] = categories

df.head()
```

### Exploratory Analysis
#### Which share of the messages is labelled with the different categories ?

```python
fig, ax = plt.subplots(figsize=(10,8))
base_color = sns.color_palette()[0]

cat_share = (df.iloc[:, 4:].sum()/df.shape[0]).sort_values(ascending=False)
sns.barplot(x=cat_share.values, y=cat_share.index, color=base_color)

ax.set_title('\n Share of messages labelled by categories \n', fontsize=14)
ax.set_xlabel('');
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{}%".format(int(100*x))))
```

**Conclusions**

Label occurence is highly skewed. More than 75% of all messages are labelled with the category 'related' whereas for several categories only very few messages are labelled.
This makes the training of a classification algorithm pretty challenging.

#### How many categories are messages allocated to ?

```python
message_category_counts = df.iloc[:, 4:].sum(axis=1)

print("Maximum number of categories a message is allocated to: ", message_category_counts.max())
print('\nMaximum number occurs for following messages:')
print(df[df.iloc[:, 4:].sum(axis=1) == message_category_counts.max()]['message'])
print('\n')
print("Minimum number of categories a message is allocated to: ", message_category_counts.min())

plt.figure(figsize=(16,6))
plt.hist( message_category_counts, bins= message_category_counts.max())
plt.title('\n Histogram on number of categories messages are allocated to \n')
plt.xlabel('Number of categories')
plt.ylabel('counts');
```

**Conclusions:**

- Most messages are not allocated to any category. 
- If catgories are allocated, 1 and 3-5 are the most 

#### Which categories co-occur with each other ?

```python
# https://stackoverflow.com/questions/20574257/constructing-a-co-occurrence-matrix-in-python-pandas
labels = df.iloc[:,4:].columns.tolist()
data = df.iloc[:,4:]

# Compute cooccurrence matrix 
cooccurrence_matrix = np.dot(data.transpose(),data)

# Compute cooccurrence matrix in percentage
# FYI: http://stackoverflow.com/questions/19602187/numpy-divide-each-row-by-a-vector-element
#      http://stackoverflow.com/questions/26248654/numpy-return-0-with-divide-by-zero/32106804#32106804
cooccurrence_matrix_diagonal = np.diagonal(cooccurrence_matrix)
with np.errstate(divide='ignore', invalid='ignore'):
    cooccurrence_matrix_percentage = np.nan_to_num(np.true_divide(cooccurrence_matrix, cooccurrence_matrix_diagonal[:, None]))

plt.figure(figsize=(16,10))
sns.heatmap(cooccurrence_matrix_percentage, xticklabels=labels, yticklabels = labels, annot=True, fmt=".1f", cmap='Blues');
```
